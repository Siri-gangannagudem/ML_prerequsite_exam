1. 
In a matrix multiplication, we need to multiply each element of a row from the first matrix to the corresponding element of a column from the second matrix. Hence in the resulting matrix, each element is the sum of k products(as we have k elements in a row in first matrix and k elements in a column in the second matrix). We get k multiplications and k-1 additions for a single element in the resulting matrix. Since total elements in the resulting matrix are nm, Total number of multiplications = nmk Total number of additions = nm*(k-1)


```python
#2
import timeit
import numpy as np

def matrix_multiply_lists(a, b):
    result = [[0 for _ in range(len(b[0]))] for _ in range(len(a))]
    
    for i in range(len(a)):
        for j in range(len(b[0])):
            for k in range(len(b)):
                result[i][j] += a[i][k] * b[k][j]
    
    return result

# Example matrices
# matrix_a = [[1, 2, 3], [4, 5, 6]]
# matrix_b = [[7, 8], [9, 10], [11, 12]]
matrix_a = np.random.rand(50, 40)
matrix_b = np.random.rand(40, 60)

# Using lists of lists
start_time_lists = timeit.default_timer()
result_lists = matrix_multiply_lists(matrix_a, matrix_b)
elapsed_time_lists = timeit.default_timer() - start_time_lists

# print("Result using lists of lists:")
# print(result_lists)
print("Time taken using lists of lists:", elapsed_time_lists, "seconds\n")

# Using NumPy
matrix_a_np = np.array(matrix_a)
matrix_b_np = np.array(matrix_b)

start_time_numpy = timeit.default_timer()
result_numpy = np.dot(matrix_a_np, matrix_b_np)
elapsed_time_numpy = timeit.default_timer() - start_time_numpy

# print("Result using NumPy:")
# print(result_numpy)
print("Time taken using NumPy:", elapsed_time_numpy, "seconds")
if elapsed_time_numpy > elapsed_time_lists:
  print("Using lists is faster")
else:
  print("Numpy is faster")
```

4. 
The gradient of $x^{2}y + y^{3}sin{x}$ with respect to $x$ is $2{x}y + y^{3}cos{x}$.
The gradient of $x^{2}y + y^{3}sin{x}$ with respect to $y$ is $x^{2} + 3{y^{2}}sin{x}$.


```python
#5
import jax
import jax.numpy as jnp

def f(x, y):
    return x**2 * y + y**3 * jnp.sin(x)

# Define the partial derivatives
def partial_derivative_x(x, y):
    return 2 * x * y + y**3 * jnp.cos(x)

def partial_derivative_y(x, y):
    return x**2 + 3* y**2 * jnp.sin(x)

# Use JAX to calculate the gradients
grad_f = jax.grad(f, argnums=(0, 1))

# Choose some random values for x and y
x_value = 2.0
y_value = 3.0

# Evaluate the analytical solution
analytical_gradient = (partial_derivative_x(x_value, y_value), partial_derivative_y(x_value, y_value))

# Evaluate the gradient using JAX
jax_gradient = grad_f(x_value, y_value)

# Print the results
print("Analytical Gradient:", analytical_gradient)
print("JAX Gradient:", jax_gradient)
```

```python
#8
class Student:
    def __init__(self, name, roll_number, marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks

class Branch:
    def __init__(self, branch_name, students):
        self.branch_name = branch_name
        self.students = students

class Year:
    def __init__(self, year, branches):
        self.year = year
        self.branches = branches

# Create instances for each student
student1 = Student('Nick', 1, {'Maths': 100, 'English': 70})
student2 = Student('Rohit', 2, {'Maths': 90, 'English': 80})
student3 = Student('Siri', 1, {'Maths': 100, 'English': 90})
student4 = Student('Akki', 2, {'Maths': 70, 'English': 70})

# Create instances for each branch
branch1 = Branch('Branch 1', [student1, student2])
branch2 = Branch('Branch 2', [student3, student4])

# Create instances for each year
year2022 = Year(2022, [branch1, branch2])
year2023 = Year(2023, [branch1, branch2])
year2024 = Year(2024, [branch1])
year2025 = Year(2025, [branch1, branch2])

# Create the overall database as a list of year objects
overall_database = [year2022, year2023, year2024, year2025]

# Example: Accessing information
maths_marks = overall_database[0].branches[0].students[0].marks['Maths']
print(f'Maths marks: {maths_marks}')

```

```python
#11
import numpy as np
import pandas as pd

np.random.seed(42)
matrix = np.random.uniform(1, 2, size=(20, 5))
df = pd.DataFrame(matrix, columns=['a', 'b', 'c', 'd', 'e'])
df['f'] = df[['a', 'b', 'c', 'd', 'e']].sum(axis=1)
df['g'] = np.where(df['f'] < 8, 'LT8', 'GT8')
num_lt8_rows = df[df['g'] == 'LT8'].shape[0]
std_lt8 = df.loc[df['g'] == 'LT8', 'f'].std()
std_gt8 = df.loc[df['g'] == 'GT8', 'f'].std()

print("DataFrame with new columns:")
print(df)
print("\nNumber of rows where 'g' is 'LT8':", num_lt8_rows)
print("Standard deviation of 'f' for 'LT8':", std_lt8)
print("Standard deviation of 'f' for 'GT8':", std_gt8)
```

Broadcasting in Numpy allows arithmetic operations between arrays of different shapes and sizes. It describes how numpy treats arrays with different shapes.


```python
#12 example to show broadcasting
import numpy as np

array1 = np.array([[1, 2, 3],
                   [4, 5, 6]])
array2 = np.array([10, 20, 30])
result = array1 + array2
print("Array 1:")
print(array1)
print("\nArray 2:")
print(array2)
print("\nResult after broadcasting:")
print(result)
```

```python
#13
import numpy as np

def find_argmin(arr):
    min_index = 0
    min_value = arr[0]

    for i in range(1, len(arr)):
        if arr[i] < min_value:
            min_value = arr[i]
            min_index = i

    return min_index
arr = np.array([4, 2, 7, 1, 9])
result_custom = find_argmin(arr)
result_np = np.argmin(arr)
print("Custom argmin result:", result_custom)
print("NumPy argmin result:", result_np)
```
